## Matplotlib
Matplotlib is a plotting library. In this section give a brief introduction to the matplotlib.pyplot module, which provides a plotting system similar to that of MATLAB.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

# Plot the points using matplotlib
plt.plot(x, y)
```

With just a little bit of extra work we can easily plot multiple lines at once, and add a title, legend, and axis labels:

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the points using matplotlib
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Sine and Cosine functions')
plt.legend(['sin(x)', 'cos(y)'])
```

### Subplots

```python
# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2, 1, 1)

# Make the first plot
plt.plot(x, y_sin)
plt.title('sin(x)')

# Set the second subplot as active, and make the second plot.
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('cos(x)')

# Show the figure.
plt.show()
```

## Confusion matrix

Consider two vectors, one containing the true label (ground-truth) and the other containing the predictions by our algorithm on a three-class classification problem:

```python
y_actu = [2, 0, 2, 2, 0, 1, 1, 2, 2, 0, 1, 2, 1, 1]
y_pred = [0, 0, 2, 1, 0, 2, 1, 0, 2, 0, 2, 2, 1, 1]
```

There are several ways to visualise the confusion matrix using Python. [Here](https://stackoverflow.com/questions/2148543/how-to-write-a-confusion-matrix-in-python) you can check some of them (e.g. using scikit-learn, pandas,...). In the following we show a simple one using scikit-learn:

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_actu, y_pred)

print(cm)
plt.matshow(cm, cmap = 'Blues')
plt.colorbar()
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

You can also normalise the confusion matrix:

```python
cm_norm =cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
plt.matshow(cm_norm, cmap = 'Blues')
plt.colorbar()
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
print(cm_norm)
```
